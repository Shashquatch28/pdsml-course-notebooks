```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
```

```python
y = x**2
```

```python
x
```

```python
y
```

```python
#Plotting
plt.plot(x, y, 'b-')

#Adding Axes 
plt.xlabel('X Label')
plt.ylabel('Y Label')

#Adding title
plt.title('Title')
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y)

plt.subplot(1, 2, 2)
plt.plot(y, x)
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y)
axes.set_xlabel('X Axis')
axes.set_ylabel('Y Axis')
axes.set_title('Plot')
plt.show()
```

```python
fig = plt.figure()

axes_1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes_2 = fig.add_axes([0.2, 0.5, 0.3, 0.3])

axes_1.plot(x, y)
axes_1.set_xlabel('X-Axis')
axes_1.set_ylabel('Y-Axis')
axes_1.set_title('Big Plot')

axes_2.plot(x, y)
axes_2.set_xlabel('x-Axis')
axes_2.set_ylabel('y-Axis')
axes_2.set_title('Small Plot')


```

```python
fig, axes = plt.subplots(1, 2)

axes[0].plot(x,y)
axes[0].set_title('Plot 1')
axes[1].plot(y, x)
axes[1].set_title('Plot 2')

plt.tight_layout()
```

```python
fig = plt.figure(figsize=(8, 2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x, y)
```

```python
fig = plt.figure(figsize = (4,4))
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x,x, label = 'x-x')
ax.plot(x, x**2, label = 'x-x**2')
ax.set_xlabel('Ordinate')
ax.set_ylabel('Abscissa')
ax.set_title('Plot')

ax.legend(loc = 0)
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])

axes.plot(x, x+1, color = 'blue', lw = 1, ls = '-', marker = 'o')
axes.plot(x, x+2, color = 'yellow', lw = 2, ls = '--', marker = 'o')
axes.plot(x, x+3, color = 'green', lw = 3, ls = '-.', marker = 'o')
```

```python
theta = np.linspace(0, 2 * np.pi, 100)
x = 16 * ( np.sin(theta) ** 3 )
y = 13 * np.cos(theta) - 5* np.cos(2*theta) - 2 * np.cos(3*theta) - np.cos(4*theta)

fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, y, 'black', lw = 2)
axes.fill(x, y, color = '#F4C2C2')
plt.text(2,1, 'F U')
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])

axes.plot(x, y, ls = '--')
axes.set_xlim(0,2)
axes.set_ylim(0,2)
```
