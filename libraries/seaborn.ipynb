# Categorical Plots 

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline 

tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.barplot(x = 'sex', y = 'total_bill', data = tips, estimator=np.std)
plt.show()
```

```python
sns.countplot(x = 'sex', data = tips)
plt.show()
```

```python
sns.boxplot(x = 'day', y = 'total_bill', data = tips)
plt.show()
```

```python
sns.violinplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex', split = True)
plt.show()
```

```python
sns.swarmplot(x = 'day', y= 'total_bill', data = tips, palette = 'coolwarm', hue = 'day')
plt.show()
```

# Matrix Plots 


```python
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline

tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
tips.head()
```

```python
flights.head()
```

```python
tc = tips.corr(numeric_only =  True)
sns.heatmap(tc, annot = True, cmap = 'mako')
plt.show()
```

```python
fp = flights.pivot_table(values = 'passengers', index = 'month', columns = 'year', observed = False)
sns.heatmap(fp, cmap = 'magma', linecolor = 'white', linewidths = 1)
plt.show()
```

```python
sns.clustermap(fp, standard_scale=1, figsize = (6,6))
plt.show()
```

# Grid Plots



```python
import seaborn as sns 
%matplotlib inline 
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.histplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
plt.show()
```

```python
g = sns.FacetGrid(data = tips, col ='time' , row = 'smoker')
g.map(sns.histplot, 'total_bill')
plt.show()
```

# Regression Plots 


```python
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```
