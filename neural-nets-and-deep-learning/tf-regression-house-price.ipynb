# Feature Engineering

```python
# drop id column since it is redundant i.e. it proveied no real value to us
df = df.drop('id', axis = 1)
```

```python
# convert string to date time object in order to extract year and month
df['date'] = pd.to_datetime(df['date'])
```

```python
df['year'] = df['date'].apply(lambda date: date.year)
df['month'] = df['date'].apply(lambda date: date.month)
```

From this plot, its hard to deduce any difference between months

# Data Preprocessing

```python
# seperate target class from features
X = df.drop('price', axis = 1).to_numpy()
y = df['price'].to_numpy()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.fit_transform(X_test)
```

```python
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense
```

```python
X_train.shape
```

```python
model = Sequential()

model.add(Dense(units = 19, activation = 'relu'))
model.add(Dense(units = 19, activation = 'relu'))
model.add(Dense(units = 19, activation = 'relu'))
model.add(Dense(units = 19, activation = 'relu'))

model.add(Dense(units = 1))

model.compile(optimizer = 'adam', loss = 'mse')
```

model is checked against test data after every epoch iteration, keras does not adjust weights and biases as per test data. This allows us to identify correct value of epoch. Since the dataset is quite large, batch size helps to decrease computation time. Smaller batch size implies more computation time but less probability of overfitting.

```python
model.fit(x = X_train, y = y_train, validation_data=(X_test, y_test), batch_size=128, epochs = 400, verbose = 2)
```

```python
loss_hist = pd.DataFrame(model.history.history)
```

```python
loss_hist.plot(figsize = (14, 5))
plt.show()
```

Since both validation loss and loss keep on decreasing, this implies that we could have continued training. If the validation loss had started spiking, that would have implied over fitting and we could have used early stopping to stop training.

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, explained_variance_score
```

```python
prediction = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, prediction)
mse
```

```python
rmse = mse**0.5
rmse
```

```python
mae = mean_absolute_error(y_test, prediction)
mae
```
